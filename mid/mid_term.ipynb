This project shows useage of LSTM model for belt convertor for store use case porpuses. It shows lstm analysis for each item on top of belt and regarding to this controls belt sequnce or speed. In real world scinario each item has its own weight, thereffore our data.csv gives information about weights. (Relying on weight sequnce will be automated)

```python
import numpy as np

def generate_sequential_data(num_sequences, sequence_length, failure_probability):
    data = []
    for _ in range(num_sequences):
        sequence = []
        for _ in range(sequence_length):
            temperature = np.random.normal(25, 5)  # Simulate temperature readings
            vibration = np.random.normal(0.1, 0.02)  # Simulate vibration readings
            belt_speed = np.random.uniform(0.5, 2.0)  # Simulate belt speed readings

            # Introduce failure event based on failure_probability
            is_failure = np.random.rand() < failure_probability

            if is_failure:
                # Simulate sensor readings during a failure event
                temperature += np.random.normal(20, 10)
                vibration += np.random.normal(0.5, 0.1)
                belt_speed -= np.random.uniform(0.1, 0.5)

            sequence.append([temperature, vibration, belt_speed, int(is_failure)])

        data.append(sequence)

    return data

# Example usage:
num_sequences = 100
sequence_length = 50
failure_probability = 0.1  # Adjust this probability as needed
#Example out put
dataset = generate_sequential_data(num_sequences, sequence_length, failure_probability)
print(dataset)
```

```python
#To check weight of each item in store
import pandas
df = pandas.read_csv("data.csv")
for item in df['Weight_in_gms']:
  print(item,'g')
```

```python
#There will be three speed to control sequnce of converter belt
import matplotlib.pyplot as plt
import numpy as np
slow = 20
average = 50
fast = 100
light_w = 0
average_w = 0
heavy_w = 0
data = []
for item_w in df['Weight_in_gms']:
  if item_w <= 3000:
    dataset = generate_sequential_data(num_sequences=fast, sequence_length=1, failure_probability=0.1)
    light_w +=1
    data.append(dataset)
  elif item_w > 3000 and item_w < 10000:
    dataset = generate_sequential_data(num_sequences=fast, sequence_length=1, failure_probability=0.1)
    average_w += 1
    data.append(dataset)
  else:
    dataset = generate_sequential_data(num_sequences=slow, sequence_length=1, failure_probability=0.1)
    heavy_w +=1
    data.append(dataset)
print("Today:")
print(f"Light box:{light_w}")
print(f"Average box:{average_w}")
print(f"Heavy box:{heavy_w}")
#To get belt converter statics for each sequnce print this however this won't work
#Because Colab limitation you need to run it in your own Jupiter enveroment
#print(data)
```
